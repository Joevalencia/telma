Este Ejemplo es dedicado al calculo de operaciones con las PERM/PASEM segun el curso de matematicas del seguro por la Universitat de Barcelona.<br>
En particular se enfoca en:
* Rentas de supervivencia no fraccionadas. O sea con misma frecuencia de variaciòn de pago y cuantìa.
* Rentas de supervivencia fraccionadas. O sea con frecuencia dee variaciòn de pago y cuantìa diferente. 
* Seguro de Fallecimiento (fraccionado y no fraccionado). Evalua Tanto el caso continuo que discreto.
* Ademas algunas operaciones conjuntas extraidas del libro "Estadistica Actuarial Vida".

---
* **SpainOperation** Realiza càlculos con el uso de las PERM/ PASEM o sea operaciones exclusivamente del estado español.
* PERM : Tablas Generacionales. $\rightarrow $ Rentas
* PASEM: Tablas no Generacionales.  $\rightarrow $ Seguros

#### Rentas de supervivencia (caso NO fraccionado)

#### Notese que si no se especifica $n$ la temporalidadse entiende vitalicia, asimismo si no se especifica $m$ diferimiento se entiende inmediata

##### $h$ es el fraccionamiento $h=1$ annual

```python
# Ejemplo 1 con la libreria telma
SpainOperation().renta_constante(data, x=45, birth_year=1974, capital=5000, m=25, interest=.02, h=1)
```

### Mismo caso pospagable

```python
## Mismo ejemplo ejercicio 1. Se especifica parametro si para renta pospagable

SpainOperation().renta_constante(data, x=45, birth_year=1974, capital=5000, m=25, interest=.02, h=1, pospagable='si')
```
